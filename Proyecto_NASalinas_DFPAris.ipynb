# Proyecto: Análisis COVID-19 en la Economia mundial.

> Este proyecto pretende analizar el efecto del COVID-19 en las economías mundiales, tomando como factor en que punto se tomaron medidas para contener la curva de contagios, así como paises donde le dan importancia nula. Con datos históricos de los índices **(MIB 30 ,IBEX35 ,DJEUROSTOXX50,FTSE100,S&P 500,DOW JONES,IPC, NIKKEI 225, KOSPI).** y series de tiempo, simular los escenarios antes y despues de la pandemia. Y con esta forma poder hacer un contraste y simular el impacto que podria tener en México en distintos casos.

En pandemias pasadas ya se ha visto un comportamiento similar en las indices bursatiles. ### Definición del problema.

> Una de las caídas más abruptas en la historia del petróleo crudo, ocurrió hace apenas seis años cuando una fuerte caída de los precios costó 200,000 empleos, casi la mitad de la fuerza laboral en estados unidos. Y ahora, la propagación del coronavirus junto con una guerra de precios del petróleo entre Rusia y Arabia Saudita amenaza con devastar una vez más la industria de servicios petroleros y sus trabajadores, y como consecuencia, en una economia basada en combustibles fósiles.

> El 20 de febrero del 2020 ocurrio el principio del colapso del mercado de valores 2020, aun no ha terminado, ya que los mercados siguen volátiles, tras la caida de los precios del petróleo, a su nivel mas bajo desde 1991; el riesgo de una recesión es prácticamente 100%.

> La guerra de precios fue provocada por una ruptura en el diálogo entre la Organización de Países Exportadores de Petróleo y Rusia sobre los recortes de producción de petróleo propuestos en en medio de la pandemia de coronavirus 2019-20. Los precios del petróleo ya habían caído un 30% desde el comienzo del año debido a una caída en la demanda. La caída de los precios fue una de las causas del colapso del mercado de valores global el 9 de marzo de 2020, conocido coloquialmente como Lunes negro.

> La reducción en la demanda de viajes y la falta de actividad fabril debido al brote impactaron significativamente la demanda de petróleo, haciendo que su precio cayera. A mediados de febrero, la Agencia Internacional de Energía pronosticó que el crecimiento de la demanda de petróleo en 2020 sería el más pequeño desde 2011. La caída de la demanda china resultó en una reunión de la Organización de Países Exportadores de Petróleo (OPEP) para discutir un posible recorte en la producción para equilibrar el pérdida de demanda.

> Casi 80 paises han acudido al FMI para financiamiento de eemergencia, debido a los "lockdowns" que han frenado la economía global. El FMI espera una importante recesión este año, pero si actuamos apropiadamente, manejando los recursos y estimulando la economía, se podria recuperar en el siguiente año. Los gobiernos ya están reduciendo las tasas de interés para tratar de estabilizar las economías que sufren el impacto de los consumidores en cuarentena y las cadenas de suministro interrumpidas.
